Em um planeta distante, astronautas lançam uma pedra no ar. Com a ajuda de uma câmera que tira fotos a intervalos de tempo regulares, eles registram a altura da pedra como função do tempo, obtendo os dados mostrados na tabela abaixo. 

```julia
t_i = collect(0:0.25:5);
y_i = [5.0, 5.75, 6.4, 6.94, 7.38, 7.72, 7.96, 8.1, 8.13, 8.07, 7.9, 7.62, 7.25, 6.77, 6.2, 5.52, 4.73, 3.85, 2.86, 1.77, 0.58];
```

```julia
using Plots

plot(t_i, y_i, 
label = "Tragetória",
xaxis = "Tempo (s)",
yaxis = "Distância (m)",
c = [:viridis :plasma :magma :inferno],
linez = x,
colorbar = false,
left_margin = 5Plots.mm,
bottom_margin = 4Plots.mm)
```
